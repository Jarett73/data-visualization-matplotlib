## Data Visualization with Matplotlib

Data Visualization is the art of representing the raw data in the form of beautiful graphs. 

```python
from matplotlib import pyplot as plt
import numpy as np
```

### Line plot

```python
# Line plot

x = np.arange(1,10,0.1)
y = 2*x+5
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Customizing Line plot

x = np.arange(1,10,0.1)
y = 2*x+5
plt.plot(x,y,color='red')
plt.title('Line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Adding 2 plots in same figure

x = np.arange(1,10,0.1)
y1 = 2*x+5
y2 = 3*x+10

plt.subplot(1,2,1) # 1-row, 2-columns, 1-activating first subplot
plt.plot(x,y1)

plt.subplot(1,2,2) # 1-row, 2-columns, 2-activating second subplot
plt.plot(x,y2)

plt.show()
```

### Bar plot

```python
# Bar plot
# used for categorical value

fruit = {'apple': 30, 'mango': 45, 'banana': 10}
names = list(fruit.keys())
quantity = list(fruit.values())
```

```python
names,quantity
```

```python
plt.barh(names, quantity, color='grey')
plt.title('Distribution of fruits')
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show()
```

### Scatter plot

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,4,3,2,9]
b = [5,6,7,8,3,2,1,4,9]

# plt.figure(figsize=(10,10)) #change the size of the figure
plt.scatter(x,a,s=100)
plt.scatter(x,b,s=150,marker='2') # s-> change the size, marker-> change the shape
plt.legend(['a','b'])  # legend is used to define the type on the graph
plt.title('X vs a&b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Histogram

```python
# Histogram 
# used for numerical values

data = [1,1,1,5,6,3,0,2,7,3,9,1,7,5,4,3,1,1,5]

plt.hist(data) 
# plt.hist(data, color='red', bins=50) 
plt.show()
```

### Box plot

```python
# Box plot

one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = list([one,two,three])
data
```

```python
plt.boxplot(data)
plt.show()
```

### Violin plot

```python
# Violin plot

one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = list([one,two,three])
plt.violinplot(data, showmedians=True, showmeans=True)
plt.grid(True) # adds a grid
plt.title("Distribution of data")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Pie chart

```python
# Pie chart

fruits = ['apple','mango','orange','grapes']
quantity = [30,45,12,100]

plt.pie(quantity, labels=fruits, autopct='%0.1f%%', shadow=True, explode=(0,0,0,0.1))
plt.show()
```

### Doughnut chart

```python
# Doughnut chart

fruits = ['apple','mango','orange','grapes']
quantity = [30,45,12,100]

plt.pie(quantity, labels=fruits, autopct='%0.1f%%', radius=2)
plt.pie([5],colors='white', radius=1)   # Note: radius should be half the size
plt.show()
```

### Area plot

```python
# Area plot

x = [10,20,30,40,50,60,70,80,90]
y = [1,2,3,5,6,4,7,4,6]

plt.stackplot(x,y)
plt.show()
```
